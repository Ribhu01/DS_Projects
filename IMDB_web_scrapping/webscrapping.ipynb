```python
import pandas as pd
import requests 
from bs4 import BeautifulSoup
from selenium import webdriver
import time
from urllib.parse import urljoin
import concurrent.futures
```

```python
url = 'https://www.imdb.com/chart/top/'
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
#webpage =requests.get(url,headers = headers).text
driver = webdriver.Chrome()
driver.get(url)


scroll_pause_time = 2
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(scroll_pause_time)
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
```

```python
soup=BeautifulSoup(driver.page_source, 'lxml')
```

```python
print(soup.prettify())
```

```python
# creating data frame to store out data
data = pd.DataFrame()
```

**Scraping movie titles of top 250 movies**

```python
scraped_movies = soup.find_all('div',class_="ipc-title ipc-title--base ipc-title--title ipc-title-link-no-icon ipc-title--on-textPrimary sc-ab348ad5-9 cIVpMc cli-title")

scraped_movies

```

```python
movies_title=[]
for movie in scraped_movies:
    movies_title.append(movie.text.strip())
movies_title
```

**Scraping movie rleased year**

```python
scraped_release_year = soup.find_all('div',class_="sc-ab348ad5-7 cqgETV cli-title-metadata")
scraped_release_year
```

```python
movies_release_year=[]
release_year=[]
for year in scraped_release_year:
    movies_release_year.append(year.text)
    
for i in movies_release_year:
    release_year.append(i[:4].strip())
release_year
```

 **Scraping Rating**

```python
rating= soup.find_all('span',class_='ipc-rating-star--rating')
imdb_rating=[]
for i in rating:
    imdb_rating.append(i.text)
imdb_rating
```

**Scraping runtime of the movie**

```python
scraped_runtime = soup.find_all('div',class_="sc-ab348ad5-7 cqgETV cli-title-metadata")
```

```python
s_data=[]
runtime=[]
for time in scraped_runtime:
    s_data.append(time.text)
for time in s_data:
    start=4
    end_index = time.find('m', start)  # Find 'm' after the start index
    if end_index != -1:
        runtime.append(time[start:end_index].strip())
    else:
        runtime.append("null")
runtime
```

```python
runtime_in_minutes = []
for time in runtime:
    if 'h' in time:
        hours, minutes = time.split('h')  
        hours = int(hours.strip())  
        minutes = int(minutes.strip())  
        runtime_in_minutes.append(hours * 60 + minutes) 
    elif time=='null':
        runtime_in_minutes.append(0)
    elif 'h' not in time:
        runtime_in_minutes.append(minutes)
len(runtime_in_minutes)
```

**Scraping certificates,genre meta-score,directors of the movies**

```python
links = soup.find_all('a',class_="ipc-title-link-wrapper")
hrefs = []
for link in links:
    href = link.get('href')
    if href:   
        full_url = urljoin(url, href)
        hrefs.append(full_url)
hrefs
# since we are getting 8 extra links we will remove those extra 8 links
hrefs = hrefs[:250]
hrefs
```

```python
# we have all the links of the each movie ,so now we can request certificates,genre meta-score,Director data from these webpg link for each movies
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}

def fetch_director(href):
    try:
        response = session.get(href, headers=headers, timeout=10).text
        soup = BeautifulSoup(response, 'lxml')
        return soup.find('a', class_="ipc-metadata-list-item__list-content-item ipc-metadata-list-item__list-content-item--link").text
    except (requests.exceptions.RequestException, AttributeError):
        return "N/A"

session = requests.Session()

with concurrent.futures.ThreadPoolExecutor(max_workers=10) as executor:
    directors = list(executor.map(fetch_director, hrefs))


```

**scraping certifications for all the 250 movies**

```python
s_certification=soup.find_all('div',class_="sc-ab348ad5-7 cqgETV cli-title-metadata")
s_certification
```

```python
scraped_data=[]
for i in s_certification:
    scraped_data.append(i.text)
scraped_data
```

```python
MAP=[]
for string in scraped_data:
    if 'm' in string:
         MAP.append(string.split('m')[-1])
    elif 'm' not in string:
        MAP.append(string.split('h')[-1])
# formating ratings
for i in range(len(MAP)):
    if MAP[i] == '15':
        MAP[i] = '15+'
    elif MAP[i] == '18':
        MAP[i] = '18+'
    elif MAP[i] == '12':
        MAP[i] = '12+'
    elif MAP[i] == 'PG':
        MAP[i] = 'PG-13'
MAP
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}

def fetch_metascore(href):
    try:
        response = session.get(href, headers=headers, timeout=10).text
        soup = BeautifulSoup(response, 'lxml')
        return soup.find('span', class_="sc-b0901df4-0 bXIOoL metacritic-score-box").text
    except (requests.exceptions.RequestException, AttributeError):
        return "N/A"

session = requests.Session()

with concurrent.futures.ThreadPoolExecutor(max_workers=10) as executor:
    metascore = list(executor.map(fetch_metascore, hrefs))
```

```python
metascore
```

```python
# scraping genre
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}

def fetch_genre(href):
    try:
        response = session.get(href, headers=headers, timeout=10).text
        soup = BeautifulSoup(response, 'lxml')
        s_genre = soup.find('div',class_="ipc-chip-list__scroller")
        genre_tag = s_genre.find_all('a')
        genre=[]
        for genre_tag in genre_tag:
            text = genre_tag.get_text(strip=True)  # Get the text and remove extra spaces
            if text:  # Only print if there's actual text
                genre.append(text)
        return genre
    except (requests.exceptions.RequestException, AttributeError):
        return "N/A"

session = requests.Session()

with concurrent.futures.ThreadPoolExecutor(max_workers=10) as executor:
    genre = list(executor.map(fetch_genre, hrefs))
genre
```
